##  "==" vs. "is"


`==` compares values
→ Use `==` when you want to know if two objects contain the same data.

````python
a = [1, 2, 3]
b = [1, 2, 3]

if a == b:
    print("they have the same values")
````

`is` compares identity → Use is when you want to know if two variables refer to the exact same object in memory.

````python
a = [1, 2, 3]
b = [1, 2, 3]

if a is b:
    print("they're not the same object in memory")
````

The most common use is checking against None:


````python
model = None

if model is None:
    print("No model loaded")
else:
    print("Model loaded!")


````

Rule of thumb:
- ✅ Use == to compare values.
- ✅ Use is when checking for None.
- ❌ Don't use is to compare numbers, strings, or lists—you'll often get surprising results because is is about identity, not equality.

## Further examples with for loop

````python
# Iterating Through Lists Over Their Index


fruits = ["apple", "banana", "cherry", "date"]

for index, fruit_name in enumerate(fruits, start=1):
    print(index, fruit_name)
````

````python
# Iterate over the keys of a dictionary

user_details = {
    "name": "alice",
    "age": 30,
    "city": "Paris"
}

for key in user_details:
    print(user_details[key])
````

````python
# Iterating Over Values: `for value in dict.values()`

for value in user_details.values():
    print(value)
````

````python
# Iterating Over Key-Value Pairs: `for key, value in dict.items()`

for key, value in user_details.items():
    # print(key, value)
    print(f"{key}... {value}")
````
